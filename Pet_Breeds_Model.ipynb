# Pet Breeds Classifiers

```python
path = Path()
Image.open(path/'Images'/'dog and cat breeds.jpg')
```

```python
learn_inf = load_learner(path/'export.pkl')
btn_upload = widgets.FileUpload()
output = widgets.Output()
lbl_pred = widgets.Label()
```

```python
def on_click_classify(change):
    img = PILImage.create(btn_upload.data[-1])
    output.clear_output()
    with output: display(img.to_thumb(128,128))
    pred,pred_idx,probs = learn_inf.predict(img)
    lbl_pred.value = f'Breed: {pred} ; Probability: {probs[pred_idx]:.04f}'
    
```

```python
btn_upload.observe(on_click_classify, names='data')
```

```python
display(VBox([widgets.Label("Select a picture of your pet"),
             btn_upload, output, lbl_pred]))
```
